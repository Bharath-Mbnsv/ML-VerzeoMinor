# DATA ANALYSIS

# Representing Test Scores in Violinplot

```python
plt.figure(figsize=(20,10))
plt.subplot(1, 4, 1)
plt.title('MATH SCORES')
sns.violinplot(y='math score',data=mb,color='grey',linewidth=2)

plt.subplot(1, 4, 2)
plt.title('READING SCORES')
sns.violinplot(y='reading score',data=mb,color='orange',linewidth=2)

plt.subplot(1, 4, 3)
plt.title('WRITING SCORES')
sns.violinplot(y='writing score',data=mb,color='green',linewidth=2)
plt.show()
```

From the above three plots its shows that most of the students score in between 60-80 in Maths whereas in reading and writing most of them score in between 50-80


# Average of Test Scores

```python
print("Average math score is    : {}".format(np.mean(mb['math score'])))
print("Average reading score is : {}".format(np.mean(mb['reading score'])))
print("Average writing score is : {}".format(np.mean(mb['writing score'])))
```

Performance by students in maths is lower when compared to reading and writing.

Best performance is in reading.

# Correlation Between Test Scores

```python
mb.corr()
```

```python
sns.heatmap(mb.corr())
```

Writing score and reading score are highly correlated which tells us that if a student reads well then the student also writes well.

Math score doesn't have much correlation so it is not neccessary that if a student performs well in maths then student has to perform well in other subjects or vice-versa.

# Relation between test scores and Gender

```python
plt.rcParams['figure.facecolor'] = "#ccffda"
plt.rcParams['axes.facecolor'] = "#ccffda"

plt.figure(figsize=(16,8))
plt.subplot(1, 3, 1)
sns.barplot(x='gender',y='math score',data=mb)
plt.title('MATH SCORES')

plt.subplot(1, 3, 2)
sns.barplot(x='gender',y='reading score',data=mb)
plt.title('READING SCORES')

plt.subplot(1, 3, 3)
sns.barplot(x='gender',y='writing score',data=mb)
plt.title('WRITING SCORES')
plt.show()
```

From the first plot, we can say that boys math scores are higher than girls scores.

From the next two plots ,we can say that girls reading and writing scores are higher than boys scores.

# Relation between test scores and Test preparation course

```python
plt.rcParams['figure.facecolor'] = "#ffffe6"
plt.rcParams['axes.facecolor'] = "#ffffe6"

plt.figure(figsize=(16,8))
plt.subplot(1, 3, 1)
sns.barplot(x='test preparation course',y='math score',data=mb)
plt.title('MATH SCORES')

plt.subplot(1, 3, 2)
sns.barplot(x='test preparation course',y='reading score',data=mb)
plt.title('READING SCORES')

plt.subplot(1, 3, 3)
sns.barplot(x='test preparation course',y='writing score',data=mb)
plt.title('WRITING SCORES')
plt.show()
```

The above plot shows that if the test preparation course is completed then they can achieve higher scores in maths, reading,
writing.

# Relation between test scores and lunch

```python
plt.figure(figsize=(16,8))
plt.subplot(1, 3, 1)
sns.barplot(x='lunch',y='math score',data=mb)
plt.title('MATH SCORES')

plt.subplot(1, 3, 2)
sns.barplot(x='lunch',y='reading score',data=mb)
plt.title('READING SCORES')

plt.subplot(1, 3, 3)
sns.barplot(x='lunch',y='writing score',data=mb)
plt.title('WRITING SCORES')
plt.show()
```

From the above plot, we can say that students who scored higher in maths, reading and writing are those whose lunch is standard.

# Relation between test scores and Test preparation course,gender

```python
plt.rcParams['figure.facecolor'] = "#e6ecff"
plt.rcParams['axes.facecolor'] = "#e6ecff"

plt.figure(figsize=(16,8))
plt.subplot(1, 3, 1)
sns.barplot(x='test preparation course',y='math score',data=mb,hue='gender')
plt.title('MATH SCORES')

plt.subplot(1, 3, 2)
sns.barplot(x='test preparation course',y='reading score',data=mb,hue='gender',palette='seismic')
plt.title('READING SCORES')

plt.subplot(1, 3, 3)
sns.barplot(x='test preparation course',y='writing score',data=mb,hue='gender',palette='viridis')
plt.title('WRITING SCORES')
plt.show()
```

From the first plot we can say that math scores of boys are better irrespective of whether they completed the course or not.

From the next two plots we can see that girls scores are better in reading and writing.

From all three graphs it says that if the course is completed then we can achieve higher scores irrespective of gender.


# Relation between test scores and Test preparation course,Lunch

```python
plt.rcParams['axes.facecolor'] = "#e6ecff"
plt.rcParams['figure.facecolor'] = "#e6ecff"

plt.figure(figsize=(16,8))
plt.subplot(1, 3, 1)
sns.barplot(x='test preparation course',y='math score',data=mb,hue='lunch',palette='viridis')
plt.title('MATH SCORES')

plt.subplot(1, 3, 2)
sns.barplot(x='test preparation course',y='reading score',data=mb,hue='lunch')
plt.title('READING SCORES')

plt.subplot(1, 3, 3)
sns.barplot(x='test preparation course',y='writing score',data=mb,hue='lunch',palette='seismic')
plt.title('WRITING SCORES')
plt.show()
```

The above plot shows,In all the cases the scores are higher who are having the standard lunch and completed the test preparation course.


# Relation between Test scores and Race/Ethnicity

```python
plt.rcParams['figure.facecolor'] = "#ffffe6"
plt.rcParams['axes.facecolor'] = "#ffffe6"

plt.figure(figsize=(16,8))
plt.subplot(1, 3, 1)
plt.title('MATH SCORES')
sns.barplot(x='race/ethnicity',y='math score',data=mb,hue='gender',palette='seismic')

plt.subplot(1, 3, 2)
plt.title('READING SCORES')
sns.barplot(x='race/ethnicity',y='reading score',data=mb,hue='gender',palette='viridis')

plt.subplot(1, 3, 3)
plt.title('WRITING SCORES')
sns.barplot(x='race/ethnicity',y='writing score',data=mb,hue='gender')
plt.show()
```

The above plot shows group E students(boys and girls) have scored higher in all three subjects(maths, reading, writing) than the other groups.


# Count according to Parents level of Education

```python
plt.rcParams['figure.facecolor'] = "#e6ecff"
plt.rcParams['axes.facecolor'] = "#e6ecff"

plt.figure(figsize=(12,6))
plt.title('PARENTS LEVEL OF EDUCATION')
sns.countplot(x='parental level of education',data=mb,palette='inferno')
plt.tight_layout()
```

The above plot shows most of the parents studied in some college or had associate degree and there are very less people who had higher studies.


# Relation between Maths score and Parents level of Education

```python
plt.rcParams['figure.facecolor'] = "#ffe6f9"
plt.rcParams['axes.facecolor'] = "#ffe6f9"

plt.figure(figsize=(12,6))
plt.title('PARENTS LEVEL OF EDUCATION vs MATHS SCORE')
sns.barplot(x='parental level of education',y='math score',data=mb,palette='seismic')
plt.tight_layout()
```

From the above plot its clear that if the parental education is some college or any degree then their children tend to score better in math.

# Relation between Writing score and Parents level of Education

```python
plt.rcParams['figure.facecolor'] = "#ffe6f9"
plt.rcParams['axes.facecolor'] = "#ffe6f9"

plt.figure(figsize=(12,6))
plt.title('PARENTS LEVEL OF EDUCATION vs WRITING SCORE')
sns.barplot(x='parental level of education',y='writing score',data=mb,palette='viridis')
plt.tight_layout()
```
